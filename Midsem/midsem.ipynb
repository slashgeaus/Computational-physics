# Q1

```python
def func(x):
    return (math.log(x/2)-math.sin(5*x/2))

def d_func(x):
    return ((1/x)-math.cos(5*x/2)*2.5)

a0, b0 = lib.bracketing(1.5, 2.5, func) 
# bracketing to choose proper interval for bisection


print("[",a0,",",b0,"]")

print("Regula Falsi:")
lib.Regula_falsi(func,1.5,3.0)

print("\n","Newton Raphson:")
lib.Newton_Raphson(func,d_func,0.5)
```
